```python
# 导入代码库
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from tqdm import tqdm

# 设置基本参数
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36 Edg/127.0.0.0'}
```

# 爬取目录页（几月auction）

```python
url = 'https://pianoauctions.co.uk/past-auctions/'
```

```python
# 请求网页内容
response = requests.get(url=url,headers=headers)
# 储存文本形式
fp = response.text
```

```python
# BeautifulSoup 解析网页内容
soup = BeautifulSoup(fp,'lxml')
# 解析定位标签内容
linklist = []
titlelist = []
length = len(soup.select('.elementor-widget-container>div>ul>li'))
for i in range(0,length):
    titlelist.append(soup.select('.elementor-widget-container>div>ul>li>a>img')[i]['alt'])
    linklist.append(soup.select('.elementor-widget-container>div>ul>li>a')[i]['href'])
#     print(soup.select('.elementor-widget-container>div>ul>li>a>img')[i]['alt'])
#     print(soup.select('.elementor-widget-container>div>ul>li>a')[i]['href'])

# 将结果储存成字典的形式
auction_dict = {
    'date_title': titlelist,
    'date_link': linklist
}
print(auction_dict['date_title'])
print(auction_dict['date_link'])
```

```python
# 爬取并输出页面中的型号、Sold_For数据以及链接
length = len(soup.select('footer > div > div > div > div > div > div > a'))
print('总共有{}条拍卖数据'.format(length))
auction_title = []
sold_for = []
piano_link_with_pics = []
for i in range(0,length):
    auction_title.append(soup.select('footer > div > div > div > div > div > div > a > div > div > div > h3')[i].text)
    sold_for.append(soup.select('footer > div > div > div > div > div > div > a > div > div > div > span')[i].text)
    piano_link_with_pics.append(soup.select('footer > div > div > div > div > div > div > a')[i]['href'])
result_dic = {
    'auction_title': auction_title,
    'sold_for': sold_for,
    'piano_link_with_pics': piano_link_with_pics
}
show_results = pd.DataFrame(result_dic)
# show_results.to_excel('test.xlsx')
show_results
```

```python
description = []
for i in tqdm(range(0,length)):
    url = result_dic['piano_link_with_pics'][i]
    response = requests.get(url=url,headers=headers)
    fp = response
    # BeautifulSoup 解析网页内容
    soup = BeautifulSoup(fp.text,'lxml')
    description.append(soup.select('.elementor-widget-container>p')[1].text)
```

```python
# 整理结果
result_dic_new = {
    'auction_title': auction_title,
    'sold_for': sold_for,
    'description': description,
    'piano_link_with_pics': piano_link_with_pics
}
show_results = pd.DataFrame(result_dic_new)
show_results
```

```python
# 整理结果2
single_results = {
    'date': [],
    'web_link': [],
    'piano_lot': [],
    'brand': [],
    'year': [],
    'estimate_low': [],
    'estimate_high': [],
    'result': [],
    'description': [],
    'piano_link_with_pics': []
}
single_results = pd.DataFrame(single_results)

for i in range(length):
    string = auction_title[i]
    string = string.replace("| Est. ", "").replace(" | Amendment", "").replace("£", "")
    string = string.split(' – ')

    piano_lot_i = string[0].replace('Lot ','')
    year_i = string[1].split('(')[-1].replace(')','')
    exclude_item = ' (' + year_i + ')'
    year_i = int(year_i.replace('c',''))
    brand_i = string[1].replace(exclude_item,'')
    estimate_low_i = int(string[2].replace(',',''))
    estimate_high_i = int(string[3].replace(',',''))

    # 处理sold_for: result
    result_i = int(sold_for[i].replace('Sold for: £','').replace(',',''))

    # 处理description: description
    description_i = description[i]

    # 处理piano_link_with_pics: piano_link_with_pics
    piano_link_with_pics_i = piano_link_with_pics[i]

    # 处理date and title: date, web_link
    web_link_i = web_link_i
    date_i = date_i.replace(' Auction Catalogue','')

    # 整理信息
    list_i = {
        'date': [date_i],
        'web_link': [web_link_i],
        'piano_lot': [piano_lot_i],
        'brand': [brand_i],
        'year': [year_i],
        'estimate_low': [estimate_low_i],
        'estimate_high': [estimate_high_i],
        'result': [result_i],
        'description': [description_i],
        'piano_link_with_pics': [piano_link_with_pics_i]
    }
    list_i = pd.DataFrame(list_i)
    single_results = single_results.append(list_i).reset_index(drop=True)
single_results.to_excel('test.xlsx')
single_results
```
